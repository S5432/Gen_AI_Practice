# **Using Basic Preprocessing Technique**

- Remove HTML tags
- Lower case
- Remove stopwords

```python
import re
def remove_tag(raw_text):
  cleaned_text = re.sub(re.compile('<.*?>'),'',raw_text)
  return cleaned_text
```

```python
df['review'] = df['review'].apply(remove_tag)
```

```python
df.head()
```

#lowercase

```python
df['review'][0]
```

```python
df['review'] =df['review'].str.lower()
```

# remove stopwords

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
sw_list = stopwords.words('english')
```

```python
len(sw_list)
```

```python
df['review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
import os
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

# Initialize an empty list to store the processed sentences
story = []

# Path to the file
file_path = '/content/IMDB Dataset.csv'

# Open the file and process its content
with open(file_path, 'r', encoding='utf-8') as f:
    corpus = f.read()  # Read the content of the file
    raw_sent = sent_tokenize(corpus)  # Tokenize the content into sentences
    for sent in raw_sent:
        story.append(simple_preprocess(sent))  # Preprocess each sentence and add to the story list

# Now 'story' will contain the preprocessed sentences
```

```python
story
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
vec = model.wv.get_normed_vectors()
```

```python
y = model.wv.index_to_key
```
